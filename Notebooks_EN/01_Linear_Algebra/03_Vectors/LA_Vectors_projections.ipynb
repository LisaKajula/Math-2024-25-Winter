# Vectors - Projections and Divisions


## Projecting a Vector onto Another Vector

Projecting vector $\mathbf{a}$ onto axis $\mathbf{b}$ involves finding the component of vector $\mathbf{a}$ along the direction of $\mathbf{b}$.

Projection of $\mathbf{a}$ onto $\mathbf{b}$:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{b}|^2} \mathbf{b}
$$


## Projecting a Vector onto Axes

Projecting vector $\mathbf{a}$ onto an axis involves finding the component of vector $\mathbf{a}$ along that axis.

Projection of $\mathbf{a}$ onto the $OX$ axis:

$$
\text{Proj}_{OX} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{i}}{|\mathbf{i}|^2} \mathbf{i} = \frac{\mathbf{a} \cdot \mathbf{i}}{1} \mathbf{i} = (\mathbf{a} \cdot \mathbf{i}) \mathbf{i}
$$

Projection of $\mathbf{a}$ onto the $OY$ axis:

$$
\text{Proj}_{OY} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{j}}{|\mathbf{j}|^2} \mathbf{j} = \frac{\mathbf{a} \cdot \mathbf{j}}{1} \mathbf{j} = (\mathbf{a} \cdot \mathbf{j}) \mathbf{j}
$$

Recall that $\mathbf{i} = (1, 0)$ and $\mathbf{j} = (0, 1)$ are unit vectors along the $OX$ and $OY$ axes.


## Dividing a Line Segment in a Given Ratio

A line segment is defined by two points in space.

Line segments and vectors are closely related because a line segment connecting two points can be treated as a vector directed from one point to another.

If point \( P \) divides the line segment connecting points $A(x_1, y_1)$ and $B(x_2, y_2)$ in the ratio $m:n$, then the coordinates of point $P$ can be found using the formula:

$$
P = \left(\frac{n \times x_2 + m \times x_1}{m+n}, \frac{n \times y_2 + m \times y_1}{m+n}\right)
$$

**Example:**

Given a line segment connecting points $A(1, 2)$ and $B(3, 8)$, find point $P$ that divides the segment in the ratio 2:3.

Coordinates of point $P$:
$$
P = \left(\frac{2 \times 3 + 3 \times 1}{2+3}, \frac{2 \times 8 + 3 \times 2}{2+3}\right) = \left(\frac{9}{5}, \frac{22}{5}\right)
$$

```python
import matplotlib.pyplot as plt
import math

# Coordinates of points A and B
A = (1, 2)
B = (3, 8)

# Ratio m:n
m, n = 2, 3

# Calculate coordinates of point P that divides the segment in the ratio 2:3
P_x = (n * A[0] + m * B[0]) / (m + n)
P_y = (n * A[1] + m * B[1]) / (m + n)
P = (P_x, P_y)

# Plotting the graph
plt.figure(figsize=(6, 6))

# Plotting line segment AB
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Line Segment AB')

# Marking points A, B, and P
plt.scatter(*A, color='blue', label='A(1, 2)')
plt.scatter(*B, color='red', label='B(3, 8)')
plt.scatter(*P, color='green', label=f'P({P[0]:.2f}, {P[1]:.2f})')

# Point labels
plt.text(A[0]-0.2, A[1], 'A(1, 2)', fontsize=12, color='blue')
plt.text(B[0]+0.1, B[1], 'B(3, 8)', fontsize=12, color='red')
plt.text(P[0]+0.1, P[1], f'P({P[0]:.2f}, {P[1]:.2f})', fontsize=12, color='green')

# Adding grid and axis labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dividing a Line Segment in the Ratio 2:3')
plt.legend()

# Displaying the graph
plt.show()
```

## 1. Divide the line segment connecting points \( A(-1, 2) \) and \( B(3, -2) \) in the ratio \( 1 : 3 \). Illustrate the result on a graph.

To divide the line segment in the ratio \( 1 : 3 \), we use the section formula:

$$
P(x, y) = \left( \frac{m_1x_2 + m_2x_1}{m_1 + m_2}, \frac{m_1y_2 + m_2y_1}{m_1 + m_2} \right)
$$

Here:
- \( A(x_1, y_1) = (-1, 2) \),
- \( B(x_2, y_2) = (3, -2) \),
- \( m_1 = 1 \), \( m_2 = 3 \).

Substituting values:

$$
P(x, y) = \left( \frac{1 \cdot 3 + 3 \cdot (-1)}{1 + 3}, \frac{1 \cdot (-2) + 3 \cdot 2}{1 + 3} \right)
$$

Simplify:

$$
P(x, y) = \left( \frac{3 - 3}{4}, \frac{-2 + 6}{4} \right) = (0, 1)
$$

Thus, the dividing point is \( P(0, 1) \).

## 2. Project vector \( \mathbf{a} = (3, 4) \) onto the \( OX \) and \( OY \) axes. Illustrate the result on a graph.

### **Projection onto \( OX \):**
The projection of \( \mathbf{a} \) onto the \( OX \)-axis retains the \( x \)-component, while the \( y \)-component becomes \( 0 \):

$$
\text{Proj}_{OX}(\mathbf{a}) = (a_x, 0) = (3, 0)
$$

### **Projection onto \( OY \):**
The projection of \( \mathbf{a} \) onto the \( OY \)-axis retains the \( y \)-component, while the \( x \)-component becomes \( 0 \):

$$
\text{Proj}_{OY}(\mathbf{a}) = (0, a_y) = (0, 4)
$$

## 3. Project vector \( \mathbf{a} = (2, 3) \) onto vector \( \mathbf{b} = (1, 1) \). Illustrate the result on a graph.

The formula for the projection of \( \mathbf{a} \) onto \( \mathbf{b} \) is:

$$
\text{Proj}_{\mathbf{b}}(\mathbf{a}) = \frac{\mathbf{a} \cdot \mathbf{b}}{\mathbf{b} \cdot \mathbf{b}} \mathbf{b}
$$

Step-by-step:

1. Compute \( \mathbf{a} \cdot \mathbf{b} \):
$$
\mathbf{a} \cdot \mathbf{b} = (2)(1) + (3)(1) = 5
$$

2. Compute \( \mathbf{b} \cdot \mathbf{b} \):
$$
\mathbf{b} \cdot \mathbf{b} = (1)(1) + (1)(1) = 2
$$

3. Substitute into the formula:
$$
\text{Proj}_{\mathbf{b}}(\mathbf{a}) = \frac{5}{2} \mathbf{b} = \frac{5}{2}(1, 1) = (2.5, 2.5)
$$

## 4. Project vector \( \mathbf{b} = (1, 1) \) onto vector \( \mathbf{a} = (2, 3) \). Illustrate the result on a graph.

The formula for the projection of \( \mathbf{b} \) onto \( \mathbf{a} \) is:

$$
\text{Proj}_{\mathbf{a}}(\mathbf{b}) = \frac{\mathbf{b} \cdot \mathbf{a}}{\mathbf{a} \cdot \mathbf{a}} \mathbf{a}
$$

Step-by-step:

1. Compute \( \mathbf{b} \cdot \mathbf{a} \):
$$
\mathbf{b} \cdot \mathbf{a} = (1)(2) + (1)(3) = 5
$$

2. Compute \( \mathbf{a} \cdot \mathbf{a} \):
$$
\mathbf{a} \cdot \mathbf{a} = (2)(2) + (3)(3) = 13
$$

3. Substitute into the formula:
$$
\text{Proj}_{\mathbf{a}}(\mathbf{b}) = \frac{5}{13} \mathbf{a} = \frac{5}{13}(2, 3) = \left(\frac{10}{13}, \frac{15}{13}\right)
$$

```python
import matplotlib.pyplot as plt
import numpy as np

# Define vector a
a = np.array([3, 4])

# Projection onto OX axis
proj_OX = np.array([a[0], 0])

# Projection onto OY axis
proj_OY = np.array([0, a[1]])

# Create the plot
plt.figure(figsize=(6, 6))

# Plot vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a')

# Plot projection onto OX axis
plt.quiver(0, 0, proj_OX[0], proj_OX[1], angles='xy', scale_units='xy', scale=1, color='red', label='Proj_OX(a)')

# Plot projection onto OY axis
plt.quiver(0, 0, proj_OY[0], proj_OY[1], angles='xy', scale_units='xy', scale=1, color='green', label='Proj_OY(a)')

# Add labels and annotations
plt.text(a[0] + 0.2, a[1], 'a(3, 4)', fontsize=12, color='blue')
plt.text(proj_OX[0] + 0.2, proj_OX[1], 'Proj_OX(a)(3, 0)', fontsize=12, color='red')
plt.text(proj_OY[0] + 0.2, proj_OY[1], 'Proj_OY(a)(0, 4)', fontsize=12, color='green')

# Set plot limits and labels
plt.xlim([-1, 4])
plt.ylim([-1, 5])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Projection of Vector a onto OX and OY Axes')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define vectors a and b
a = np.array([2, 3])
b = np.array([1, 1])

# Calculate the projection of a onto b
proj_b_a = (np.dot(a, b) / np.dot(b, b)) * b

# Create the plot
plt.figure(figsize=(6, 6))

# Plot vector a
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a')

# Plot vector b
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='b')

# Plot projection of a onto b
plt.quiver(0, 0, proj_b_a[0], proj_b_a[1], angles='xy', scale_units='xy', scale=1, color='green', label='Proj_b(a)')

# Add labels and annotations
plt.text(a[0] + 0.2, a[1], 'a(2, 3)', fontsize=12, color='blue')
plt.text(b[0] + 0.2, b[1], 'b(1, 1)', fontsize=12, color='red')
plt.text(proj_b_a[0] + 0.2, proj_b_a[1], 'Proj_b(a)(2.5, 2.5)', fontsize=12, color='green')

# Set plot limits and labels
plt.xlim([-1, 4])
plt.ylim([-1, 4])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Projection of Vector a onto Vector b')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
