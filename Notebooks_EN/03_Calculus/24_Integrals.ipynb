## 1. Compute the Following Integrals

### (a) $$ \int 1 \, dx $$
$$ \int 1 \, dx = x + C $$

### (b) $$ \int (x^2 + 2) \, dx $$
$$ \int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C $$

### (c) $$ \int 2 \sin(x) \, dx $$
$$ \int 2 \sin(x) \, dx = -2 \cos(x) + C $$

### (d) $$ \int 3x \, dx $$
$$ \int 3x \, dx = \frac{3x^2}{2} + C $$

### (e) $$ \int \frac{1}{x^2} \, dx $$
$$ \int \frac{1}{x^2} \, dx = -\frac{1}{x} + C $$

### (f) $$ \int \left( \frac{1}{3}x^4 - 5 \right) \, dx $$
$$ \int \left( \frac{1}{3}x^4 - 5 \right) \, dx = \frac{x^5}{15} - 5x + C $$

### (g) $$ \int (\sin^2(x) + \cos^2(x)) \, dx $$
Since $$ \sin^2(x) + \cos^2(x) = 1 $$:
$$ \int (\sin^2(x) + \cos^2(x)) \, dx = \int 1 \, dx = x + C $$

### (h) $$ \int (5 \sin(x) + 3e^x) \, dx $$
$$ \int (5 \sin(x) + 3e^x) \, dx = -5 \cos(x) + 3e^x + C $$

### (i) $$ \int x^3 \, dx $$
$$ \int x^3 \, dx = \frac{x^4}{4} + C $$

### (j) $$ \int 10^x \, dx $$
$$ \int 10^x \, dx = \frac{10^x}{\ln(10)} + C $$

### (k) $$ \int \cos\left( \frac{5}{2}x + 3 \right) \, dx $$
Let $$ u = \frac{5}{2}x + 3 $$, then $$ du = \frac{5}{2} dx $$ or $$ dx = \frac{2}{5} du $$:
$$ \int \cos(u) \cdot \frac{2}{5} du = \frac{2}{5} \sin(u) + C = \frac{2}{5} \sin\left( \frac{5}{2}x + 3 \right) + C $$

### (l) $$ \int \frac{\cos(\ln(x))}{x} \, dx $$
Let $$ u = \ln(x) $$, then $$ du = \frac{1}{x} dx $$:
$$ \int \cos(u) \, du = \sin(u) + C = \sin(\ln(x)) + C $$

### (m) $$ \int x \ln(x) \, dx $$
Use integration by parts: Let $$ u = \ln(x) $$ and $$ dv = x \, dx $$. Then:
$$ \int x \ln(x) \, dx = \frac{x^2}{2} \ln(x) - \int \frac{x^2}{2} \cdot \frac{1}{x} \, dx = \frac{x^2}{2} \ln(x) - \frac{x^2}{4} + C $$

### (n) $$ \int x e^x \, dx $$
Use integration by parts: Let $$ u = x $$ and $$ dv = e^x \, dx $$. Then:
$$ \int x e^x \, dx = x e^x - \int e^x \, dx = x e^x - e^x + C $$

## 6. Find the Distance of a Moving Particle Between $$ t = 0 $$ and $$ t = 2 $$

- The position function is:
  $$ x(t) = 3t^2 - 6t + 1 $$

- The distance traveled is the integral of the speed (absolute value of velocity):
  $$ v(t) = \frac{dx}{dt} = 6t - 6 $$
  $$ \text{Distance} = \int_0^2 |6t - 6| \, dt $$

- Break the integral into two parts (where \( v(t) \) changes sign):
  - For \( t \in [0, 1] \): \( v(t) = 6t - 6 \leq 0 \), so \( |v(t)| = -v(t) \).
  - For \( t \in [1, 2] \): \( v(t) = 6t - 6 \geq 0 \), so \( |v(t)| = v(t) \).

- Compute:
  $$ \text{Distance} = \int_0^1 (-6t + 6) \, dt + \int_1^2 (6t - 6) \, dt $$
  $$ = \left[ -3t^2 + 6t \right]_0^1 + \left[ 3t^2 - 6t \right]_1^2 $$
  $$ = (-3 + 6) + (12 - 12 - 3 + 6) = 3 + 3 = 6 $$
