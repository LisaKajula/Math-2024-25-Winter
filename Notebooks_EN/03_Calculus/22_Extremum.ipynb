# Problem 1: Maximizing Profit

The profit function is given by:
$$ P(u) = -2u^2 + 50u - 300 $$

To find the number of units \( u \) that maximize profit, we find the vertex of the parabola (since it opens downward, the vertex is the maximum point).

1. **Find the derivative of \( P(u) \):**
   $$ P'(u) = \frac{d}{du}(-2u^2 + 50u - 300) = -4u + 50 $$

2. **Set the derivative equal to zero to find the critical point:**
   $$ -4u + 50 = 0 $$
   $$ -4u = -50 $$
   $$ u = \frac{50}{4} = 12.5 $$

3. **Conclusion:**
   The number of units that maximize profit is \( u = 12.5 \). Since we can't sell half a unit, we round to \( u = 12 \) or \( u = 13 \).

# Problem 2: Maximizing Rectangular Area with 10 Meters of String

We have 10 meters of string to enclose a rectangle. Let the dimensions of the rectangle be \( x \) (length) and \( y \) (width).

1. **Perimeter constraint:**
   $$ 2x + 2y = 10 $$
   Simplify:
   $$ x + y = 5 $$
   $$ y = 5 - x $$

2. **Area of the rectangle:**
   $$ A = x \cdot y = x(5 - x) = 5x - x^2 $$

3. **Find the derivative of \( A(x) \):**
   $$ A'(x) = \frac{d}{dx}(5x - x^2) = 5 - 2x $$

4. **Set the derivative equal to zero to find the critical point:**
   $$ 5 - 2x = 0 $$
   $$ 2x = 5 $$
   $$ x = 2.5 $$

5. **Find \( y \):**
   $$ y = 5 - x = 5 - 2.5 = 2.5 $$

6. **Conclusion:**
   The rectangle with the largest area is a square with dimensions \( x = 2.5 \) meters and \( y = 2.5 \) meters.

# Problem 3: Finding the Extremum of \( f(x) = x^2 + 3x - 5 \)

The function is:
$$ f(x) = x^2 + 3x - 5 $$

1. **Find the derivative of \( f(x) \):**
   $$ f'(x) = \frac{d}{dx}(x^2 + 3x - 5) = 2x + 3 $$

2. **Set the derivative equal to zero to find the critical point:**
   $$ 2x + 3 = 0 $$
   $$ 2x = -3 $$
   $$ x = -\frac{3}{2} = -1.5 $$

3. **Verify if it's a minimum or maximum:**
   The second derivative is:
   $$ f''(x) = 2 $$
   Since \( f''(x) > 0 \), the function has a minimum at \( x = -1.5 \).

4. **Conclusion:**
   The function \( f(x) = x^2 + 3x - 5 \) has a minimum at \( x = -1.5 \).

# Problem 4: Finding the Extremum of $ f(x) = \frac{x^2 + 2x + 1}{x - 1} $

The function is:
$$ f(x) = \frac{x^2 + 2x + 1}{x - 1} $$

1. **Simplify the function:**
   $$ f(x) = \frac{(x + 1)^2}{x - 1} $$

2. **Find the derivative of \( f(x) \) using the quotient rule:**
   $$ f'(x) = \frac{(2(x + 1)(x - 1)) - (x + 1)^2}{(x - 1)^2} $$
   Simplify:
   $$ f'(x) = \frac{(2(x^2 - 1)) - (x^2 + 2x + 1)}{(x - 1)^2} $$
   $$ f'(x) = \frac{2x^2 - 2 - x^2 - 2x - 1}{(x - 1)^2} $$
   $$ f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2} $$

3. **Set the derivative equal to zero to find critical points:**
   $$ x^2 - 2x - 3 = 0 $$
   Solve the quadratic equation:

   $$ x = 3 \quad \text{or} \quad x = -1 $$

4. **Verify if they are minima or maxima:**
   - For \( x = 3 \):
     $$ f''(3) > 0 $$ (minimum)
   - For \( x = -1 \):
     $$ f''(-1) < 0 $$ (maximum)

5. **Conclusion:**
