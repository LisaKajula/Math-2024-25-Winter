## Solving First-Order Ordinary Differential Equations  

### 1. Solve:  
1. $$y'(x) = y$$  
   Using the method of separation of variables:  
   $$\frac{dy}{y} = dx$$  
   Integrating both sides:  
   $$\ln|y| = x + C$$  
   Exponentiating:  
   $$y = Ce^x$$, where $$C$$ is the integration constant.  

2. $$y'(x) = \frac{1}{2}y(x)$$  
   Using separation of variables:  
   $$\frac{dy}{y} = \frac{1}{2}dx$$  
   Integrating both sides:  
   $$\ln|y| = \frac{x}{2} + C$$  
   Exponentiating:  
   $$y = Ce^{x/2}$$, where $$C$$ is the integration constant.  

---

### 2. Solve (Separation of Variables):  
1. $$\frac{dy}{dx} = xy$$  
   Rearrange:  
   $$\frac{1}{y}dy = x dx$$  
   Integrate:  
   $$\ln|y| = \frac{x^2}{2} + C$$  
   Exponentiate:  
   $$y = Ce^{x^2/2}$$  

2. $$\frac{dy}{dx} = \frac{y}{x}$$  
   Rearrange:  
   $$\frac{1}{y}dy = \frac{1}{x}dx$$  
   Integrate:  
   $$\ln|y| = \ln|x| + C$$  
   Simplify:  
   $$y = Cx$$  

3. $$\frac{dy}{dx} = xy$$  
   Rearrange and solve as in the first case above:  
   $$y = Ce^{x^2/2}$$

## Solving Second-Order Ordinary Differential Equations  

### 1. $$y''(x) + y'(x) = 0$$  
   Let $$y'(x) = v(x)$$. Then, $$y''(x) = v'(x)$$. The equation becomes:  
   $$v'(x) + v(x) = 0$$  
   Solve as a first-order equation:  
   $$\frac{dv}{v} = -dx$$  
   $$\ln|v| = -x + C$$  
   $$v = Ce^{-x}$$  
   Since $$v = y'(x)$$, integrate:  
   $$y = \int Ce^{-x} dx = -Ce^{-x} + D$$  
   Using boundary conditions $$y(0) = 2$$ and $$y'(0) = -1$$:  
   - $$y(0) = -C + D = 2 \implies D = 2 + C$$  
   - $$y'(0) = -C = -1 \implies C = 1$$  
   Substituting:  
   $$y = -e^{-x} + 3$$  

---

### 2. $$y''(x) - y(x) = 0$$  
   The characteristic equation is:  
   $$r^2 - 1 = 0 \implies r = \pm 1$$  
   General solution:  
   $$y(x) = C_1e^x + C_2e^{-x}$$  
   Using boundary conditions $$y(0) = 2$$ and $$y'(0) = 0$$:  
   - $$y(0) = C_1 + C_2 = 2$$  
   - $$y'(x) = C_1e^x - C_2e^{-x} \implies y'(0) = C_1 - C_2 = 0 \implies C_1 = C_2 = 1$$  
   Solution:  
   $$y(x) = e^x + e^{-x}$$  

---

### 3. $$\frac{d^2y(x)}{dx^2} = -\omega^2y(x)$$  
   The characteristic equation is:  
   $$r^2 + \omega^2 = 0 \implies r = \pm i\omega$$  
   General solution:  
   $$y(x) = C_1\cos(\omega x) + C_2\sin(\omega x)$$

## Partial Differential Equation: Wave Equation  

Given: $$\psi(t, x) = A\cos(\omega t + kx)$$  
Wave equation:  
$$\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0$$  
1. Compute $$\frac{\partial^2 \psi}{\partial t^2}$$:  
   $$\frac{\partial \psi}{\partial t} = -A\omega\sin(\omega t + kx)$$  
   $$\frac{\partial^2 \psi}{\partial t^2} = -A\omega^2\cos(\omega t + kx)$$  

2. Compute $$\frac{\partial^2 \psi}{\partial x^2}$$:  
   $$\frac{\partial \psi}{\partial x} = -Ak\sin(\omega t + kx)$$  
   $$\frac{\partial^2 \psi}{\partial x^2} = -Ak^2\cos(\omega t + kx)$$  

3. Substitute into the wave equation:  
   $$-A\omega^2\cos(\omega t + kx) - v^2(-Ak^2\cos(\omega t + kx)) = 0$$  
   Simplify:  
   $$-A\cos(\omega t + kx)(\omega^2 - v^2k^2) = 0$$  
   For this to hold, $$\omega^2 = v^2k^2$$, which implies $$v = \frac{\omega}{k}$$.  

Thus, $$\psi(t, x) = A\cos(\omega t + kx)$$ satisfies the wave equation.
