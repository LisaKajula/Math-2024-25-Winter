## Solving First-Order Ordinary Differential Equations  
## Derivatives of Functions

1. **Function**:  
   $$ y(x) = -3x + 3 $$  
   **Solution**:  
   The derivative of a linear function \( ax + b \) is simply the coefficient of \( x \), which is \( a \).  
   $$ \frac{dy}{dx} = -3 $$

2. **Function**:  
   $$ y(x) = \pi x + \sin(1) $$  
   **Solution**:  
   The derivative of $$ \pi x $$ is $$\pi$$, and the derivative of a constant \( \sin(1) \) is \( 0 \).  
   $$ \frac{dy}{dx} = \pi $$

3. **Function**:  
   $$ y(x) = 4 + \sin(2) $$  
   **Solution**:  
   Both terms are constants, so their derivative is zero.  
   $$ \frac{dy}{dx} = 0 $$

4. **Function**:  
   $$ y(x) = 2x^3 - 3x^2 + 8x - 9 $$  
   **Solution**:  
   Use the power rule:  
   $$ \frac{d}{dx} x^n = n x^{n-1} $$  
   Differentiate term by term:  
   $$ \frac{dy}{dx} = 6x^2 - 6x + 8 $$

5. **Function**:  
   $$ y(x) = 6x^{1/3} $$  
   **Solution**:  
   Use the power rule:  
   $$ \frac{d}{dx} x^n = n x^{n-1} $$  
   $$ \frac{dy}{dx} = 2x^{-2/3} $$

6. **Function**:  
   $$ y(x) = x $$  
   **Solution**:  
   The derivative of \( x \) is \( 1 \).  
   $$ \frac{dy}{dx} = 1 $$

7. **Function**:  
   $$ y(x) = \cos(x) + \sin(x) $$  
   **Solution**:  
   - The derivative of $$ \cos(x) $$ is $$\sin(x)$$.  
   - The derivative of $$\sin(x) $$ is $$ \cos(x)$$.  
   $$ \frac{dy}{dx} = -\sin(x) + \cos(x) $$

8. **Function**:  
   $$ y(x) = 2\sin(x)\cos(x) $$  
   **Solution**:  
   Use the product rule:  
   $$ \frac{d}{dx}(uv) = u'v + uv' $$  
   -  u = $$2\sin(x)$$, u' = 2\cos(x) $$
   - v = $$\cos(x)$$, v' = \sin(x)
   $$ \frac{dy}{dx} = 2\cos^2(x) - 2\sin^2(x) $$  
   Simplify using the double-angle identity:  
   $$ \frac{dy}{dx} = 2\cos(2x) $$

9. **Function**:  
   $$ y(x) = x\sin(x) $$  
   **Solution**:  
   Use the product rule:  
   $$ \frac{dy}{dx} = \sin(x) + x\cos(x) $$

10. **Function**:  
    $$ y(x) = (x+1)(x+1) $$  
    **Solution**:  
    Expand the function:  
    $$ y(x) = x^2 + 2x + 1 $$  
    Differentiate term by term:  
    $$ \frac{dy}{dx} = 2x + 2 $$

11. **Function**:  
    $$ y(x) = x^{x+1} $$  
    **Solution**:  
    Rewrite as:  
    $$ y(x) = e^{(x+1)\ln(x)} $$  
    Use the chain rule:  
    $$ \frac{dy}{dx} = x^{x+1} (\ln(x) + 1) $$

12. **Function**:  
    $$ y(x) = (x+1)\exp(x) $$  
    **Solution**:  
    Use the product rule:  
    $$ \frac{dy}{dx} = \exp(x)(x+2) $$

13. **Function**:  
    $$ y(x) = \sin(x^2) $$  
    **Solution**:  
    Use the chain rule:  
    $$ \frac{dy}{dx} = 2x\cos(x^2) $$

14. **Function**:  
    $$ y(x) = \exp(-2x) $$  
    **Solution**:  
    Use the chain rule:  
    $$ \frac{dy}{dx} = -2\exp(-2x) $$

15. **Function**:  
    $$ y(x) = \frac{1}{\sin(x+1)} $$  
    **Solution**:  
    Rewrite as:  
    $$ y(x) = (\sin(x+1))^{-1} $$  
    Use the chain rule:  
    $$ \frac{dy}{dx} = -\frac{\cos(x+1)}{\sin^2(x+1)} $$

## Proofs

1. **Proof**:  
   $$ \frac{d}{dx}(\ln(\sin(x))) = \cot(x) $$  
   Solution:  
   $$ \frac{d}{dx}(\ln(\sin(x))) = \frac{1}{\sin(x)} \cdot \cos(x) $$  
   $$ \frac{d}{dx}(\ln(\sin(x))) = \cot(x) $$

2. **Verification**:  
   $$ f(x) = \cos(x), \quad f''(x) = -f(x) $$  
   Solution:  
   $$ f'(x) = -\sin(x) $$  
   $$ f''(x) = -\cos(x) $$  
   Hence,  
   $$ f''(x) = -f(x) $$

## Limits Using de l'Hospital's Rule

1. $$ \lim_{x \to 0} \frac{\sin(x)}{x} = 1 $$  
   Solution: Apply the rule:  
   $$ \frac{\cos(x)}{1} = 1 $$  

2. $$ \lim_{x \to \infty} \frac{\ln(x)}{x} = 0 $$  
   Solution: Apply the rule:  
   $$ \frac{1/x}{1} = 0 $$
