# 1. Equation of the Plane Through Points A, B, and C
**Question:** Find the equation of the plane passing through points $$ A(1, 2, 3) $$, $$ B(3, 4, 5) $$, and $$ C(2, 1, 4) $$.

To find the equation of the plane, we need to compute two vectors on the plane and their cross product to get the normal vector.

- Vector $$ \overrightarrow{AB} = B - A = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2) $$
- Vector $$ \overrightarrow{AC} = C - A = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1) $$

Now, compute the cross product of $$ \overrightarrow{AB} $$ and $$ \overrightarrow{AC} $$:

$$
\overrightarrow{AB} \times \overrightarrow{AC} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
2 & 2 & 2 \\
1 & -1 & 1
\end{vmatrix}
$$

The result is the normal vector $$ \vec{n} = (4, -4, -4) $$.

Thus, the equation of the plane is:

$$
4(x - 1) - 4(y - 2) - 4(z - 3) = 0
$$

Simplify to get the final equation:

$$
x - y - z = -1
$$

# 4. Line of Intersection of Two Planes
**Question:** Find the line of intersection of the planes $$ 2x + 3y + 4z = 5 $$ and $$ 3x + 4y + 2z = 6 $$.

To find the line of intersection, we solve the system of equations by eliminating one variable.

1. Multiply the first equation by 3 and the second by 2:

$$
6x + 9y + 12z = 15
$$

$$
6x + 8y + 4z = 12
$$

2. Subtract the second equation from the first:

$$
(6x + 9y + 12z) - (6x + 8y + 4z) = 15 - 12
$$

$$
y + 8z = 3
$$

Thus, $$ y = 3 - 8z $$.

Substitute $$ y = 3 - 8z $$ into one of the original equations:

$$
2x + 3(3 - 8z) + 4z = 5
$$

Simplify and solve for $$ x $$:

$$
2x + 9 - 24z + 4z = 5 \implies 2x = 20z - 4 \implies x = 10z - 2
$$

The parametric equations of the line are:

$$
x = 10z - 2, \quad y = 3 - 8z, \quad z = z
$$

Thus, the parametric form of the line of intersection is:

$$
(x, y, z) = (-2, 3, 0) + z(10, -8, 1)
$$

# 7. Distance from Point A to the Plane
**Question:** Find the distance from point $$ A(1, 2, 3) $$ to the plane $$ 2x + 3y + 4z = 5 $$.

The formula for the distance from a point $$ (x_1, y_1, z_1) $$ to a plane $$ Ax + By + Cz + D = 0 $$ is:

$$
d = \frac{|Ax_1 + By_1 + Cz_1 + D|}{\sqrt{A^2 + B^2 + C^2}}
$$

For the plane $$ 2x + 3y + 4z = 5 $$, we have $$ A = 2, B = 3, C = 4, D = -5 $$ and point $$ A(1, 2, 3) $$.

Substitute into the formula:

$$
d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}} = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}} = \frac{|15|}{\sqrt{29}} = \frac{15}{\sqrt{29}}
$$

# 9. Calculate the angle between the plane $$x + y + z = 1$$ and the plane $$x = 0$$ (the $$yz$$-plane)

### Question:
Find the angle between the planes $$x + y + z = 1$$ and $$x = 0$$ (the $$yz$$-plane).

---

### Concept:
The angle between two planes is determined by the angle between their normal vectors. Let the normal vectors of the two planes be:

1. For the plane $$x + y + z = 1$$, the normal vector is:  
   $$\vec{n_1} = (1, 1, 1)$$

2. For the plane $$x = 0$$ (the $$yz$$-plane), the normal vector is:  
   $$\vec{n_2} = (1, 0, 0)$$

The angle $$\theta$$ between the two planes can be calculated using:  
$$\cos\theta = \frac{|\vec{n_1} \cdot \vec{n_2}|}{||\vec{n_1}|| \cdot ||\vec{n_2}||}$$

visualization:https://www.geogebra.org/calculator/ufusnng4

---

### Solution:
1. Compute the dot product:  
   $$\vec{n_1} \cdot \vec{n_2} = (1)(1) + (1)(0) + (1)(0) = 1$$

2. Compute the magnitudes of the normal vectors:  
   $$||\vec{n_1}|| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}$$  
   $$||\vec{n_2}|| = \sqrt{1^2 + 0^2 + 0^2} = 1$$

3. Substitute into the formula for $$\cos\theta$$:  
   $$\cos\theta = \frac{|1|}{\sqrt{3} \cdot 1} = \frac{1}{\sqrt{3}}$$

4. Calculate $$\theta$$:  
   $$\theta = \arccos\left(\frac{1}{\sqrt{3}}\right)$$

---

### Final Answer:
The angle between the two planes is:  
$$\theta = \arccos\left(\frac{1}{\sqrt{3}}\right)$$
